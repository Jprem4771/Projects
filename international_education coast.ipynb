```python
#total_cost for a person to study in USD in each country
import plotly.express as px

# Calculate total cost in USD
df['Total_Cost_USD'] = (
    df['Tuition_USD'] +
    (df['Rent_USD']) +
    df['Visa_Fee_USD'] +
    df['Insurance_USD']
)

# Group by country
country_cost = df.groupby('Country')['Total_Cost_USD'].mean().reset_index()

# Create Tree Map
fig = px.treemap(
    country_cost,
    path=['Country'],
    values='Total_Cost_USD',
    color='Total_Cost_USD',
    color_continuous_scale='RdBu',
    title='Tree Map of Average Total Education Cost by Country (USD)'
)

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
df = pd.read_csv("International_Education_Costs.csv")

# Calculate total cost in USD
df['Total_Cost_USD'] = (
    df['Tuition_USD'] +
    (df['Rent_USD'] ) +
    df['Visa_Fee_USD'] +
    df['Insurance_USD']
)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Duration_Years', y='Total_Cost_USD', hue='Level', palette='deep')
plt.title('Total Cost vs Program Duration by Education Level')
plt.xlabel('Duration (Years)')
plt.ylabel('Total Cost (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Boxplot of cost distribution per level
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Level', y='Total_Cost_USD', palette='Set2')
plt.title('Total Cost Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Total Cost (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
level_cost = df.groupby('Level')['Total_Cost_USD'].mean().reset_index().sort_values(by='Total_Cost_USD', ascending=False)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=level_cost, x='Level', y='Total_Cost_USD', palette='pastel')
plt.title('Average Total Education Cost by Level')
plt.xlabel('Education Level')
plt.ylabel('Average Total Cost (USD)')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Group data
grouped = df.groupby(['Level', 'Program']).size().reset_index(name='Count')

# Inner Pie (Level)
level_counts = grouped.groupby('Level')['Count'].sum()
inner_labels = level_counts.index.tolist()
inner_sizes = level_counts.values

# Outer Pie (Program within Level)
outer_labels = grouped['Program']
outer_sizes = grouped['Count']

# Color palettes
inner_colors = plt.cm.Pastel1.colors
outer_colors = plt.cm.Set3.colors

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('equal')

# Inner ring (Level)
inner_pie = ax.pie(
    inner_sizes,
    radius=1,
    labels=inner_labels,
    labeldistance=0.7,
    colors=inner_colors,
    wedgeprops=dict(width=0.3, edgecolor='w')
)

# Outer ring (Programs)
outer_pie = ax.pie(
    outer_sizes,
    radius=2.9,
    labels=outer_labels,
    labeldistance=1.05,
    colors=outer_colors,
    wedgeprops=dict(width=0.3, edgecolor='w')
)

plt.title('Nested Pie Chart: Education Level and Programs', fontsize=16)
plt.show()
```

```python
import plotly.express as px

# Load data
df = pd.read_csv("International_Education_Costs.csv")

# Group data: count of records by Level and Program
grouped = df.groupby(['Level', 'Program']).size().reset_index(name='Count')

# Create sunburst chart
fig = px.sunburst(
    grouped,
    path=['Level', 'Program'],  # Hierarchy: Level -> Program
    values='Count',
    color='Level',
    color_discrete_sequence=px.colors.qualitative.Pastel,
    title='Interactive Sunburst: Education Levels and Programs'
)

fig.update_layout(margin=dict(t=50, l=0, r=0, b=0))
fig.show()
```

```python

# Compute total cost in USD
df['Total_Cost_USD'] = (
    df['Tuition_USD'] +
    (df['Rent_USD']) +
    df['Visa_Fee_USD'] +
    df['Insurance_USD']
)

# Group by university and get mean cost and country
uni_costs = df.groupby(['University', 'Country'])['Total_Cost_USD'].mean().reset_index()

# Get top 10 most expensive universities
top10 = uni_costs.sort_values(by='Total_Cost_USD', ascending=False).head(10)

# Plot
plt.figure(figsize=(14, 7))
sns.barplot(data=top10, y='University', x='Total_Cost_USD', hue='Country', dodge=False, palette='tab10')
plt.title('Top 10 Most Expensive Universities and Their Countries')
plt.xlabel('Average Total Cost (USD)')
plt.ylabel('University')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True, axis='x')
plt.show()
```
